```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
def init_weights():
    W1 = np.random.randn(2,2)
    b1 = np.zeros((1,2))
    W2 = np.random.randn(2,1)
    b2 = np.zeros((1,1))
    return W1, b1, W2, b2

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_deriv(x):
    return x*(1-x)
```

```python
def forward(X, W1, b1, W2, b2):
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)
    return z1, a1, z2, a2

def backward(X, y, z1, a1, z2, a2, W2):
    m = X.shape[0]

    dz2 = a2 - y
    dW2 = np.dot(a1.T, dz2)/m
    db2 = np.sum(dz2)/m

    dz1 = np.dot(dz2, W2.T) * sigmoid_deriv(a1)
    dW1 = np.dot(X.T, dz1)/m
    db1 = np.sum(dz1)/m

    return dW1, db1, dW2, db2
```

1. Batch Gradient Descent (BGD)

```python
def train_bgd(lr=0.1, epochs=10000):
    W1, b1, W2, b2 = init_weights()

    for i in range(epochs):
        z1, a1, z2, a2 = forward(X, W1, b1, W2, b2)
        dW1, db1, dW2, db2 = backward(X, y, z1, a1, z2, a2, W2)

        W1 -= lr * dW1
        b1 -= lr * db1
        W2 -= lr * dW2
        b2 -= lr * db2

    return W1, b1, W2, b2
```

2. Stochastic Gradient Descent (SGD)

```python
def train_sgd(lr=0.1, epochs=10000):
    W1, b1, W2, b2 = init_weights()

    for i in range(epochs):
        for j in range(len(X)):
            xi = X[j:j+1]
            yi = y[j:j+1]

            z1, a1, z2, a2 = forward(xi, W1, b1, W2, b2)
            dW1, db1, dW2, db2 = backward(xi, yi, z1, a1, z2, a2, W2)

            W1 -= lr * dW1
            b1 -= lr * db1
            W2 -= lr * dW2
            b2 -= lr * db2

    return W1, b1, W2, b2
```

3. Mini-Batch GD

```python
def train_minibatch(lr=0.1, epochs=10000, batch_size=2):
    W1, b1, W2, b2 = init_weights()

    for i in range(epochs):
        for j in range(0, len(X), batch_size):
            xb = X[j:j+batch_size]
            yb = y[j:j+batch_size]

            z1, a1, z2, a2 = forward(xb, W1, b1, W2, b2)
            dW1, db1, dW2, db2 = backward(xb, yb, z1, a1, z2, a2, W2)

            W1 -= lr * dW1
            b1 -= lr * db1
            W2 -= lr * dW2
            b2 -= lr * db2

    return W1, b1, W2, b2
```

4. SGD with Momentum

```python
def train_momentum(lr=0.1, beta=0.9, epochs=10000):
    W1, b1, W2, b2 = init_weights()
    vW1 = vW2 = vb1 = vb2 = 0

    for i in range(epochs):
        z1, a1, z2, a2 = forward(X, W1, b1, W2, b2)
        dW1, db1, dW2, db2 = backward(X, y, z1, a1, z2, a2, W2)

        vW1 = beta*vW1 + (1-beta)*dW1
        vW2 = beta*vW2 + (1-beta)*dW2

        W1 -= lr * vW1
        W2 -= lr * vW2

    return W1, b1, W2, b2
```

6. Adagrad

```python
def train_adagrad(lr=0.1, epochs=10000):
    W1, b1, W2, b2 = init_weights()
    eps = 1e-8
    G1 = G2 = 0

    for i in range(epochs):
        z1, a1, z2, a2 = forward(X, W1, b1, W2, b2)
        dW1, db1, dW2, db2 = backward(X, y, z1, a1, z2, a2, W2)

        G1 += dW1**2
        G2 += dW2**2

        W1 -= lr * dW1 / np.sqrt(G1 + eps)
        W2 -= lr * dW2 / np.sqrt(G2 + eps)

    return W1, b1, W2, b2
```

7. RMSProp

```python
def train_rmsprop(lr=0.01, beta=0.9, epochs=10000):
    W1, b1, W2, b2 = init_weights()
    S1 = S2 = 0
    eps = 1e-8

    for i in range(epochs):
        z1, a1, z2, a2 = forward(X, W1, b1, W2, b2)
        dW1, db1, dW2, db2 = backward(X, y, z1, a1, z2, a2, W2)

        S1 = beta*S1 + (1-beta)*(dW1**2)
        S2 = beta*S2 + (1-beta)*(dW2**2)

        W1 -= lr * dW1 / np.sqrt(S1 + eps)
        W2 -= lr * dW2 / np.sqrt(S2 + eps)

    return W1, b1, W2, b2
```

8. Adam (BEST)

```python
def train_adam(lr=0.01, epochs=10000):
    W1, b1, W2, b2 = init_weights()

    m1 = m2 = v1 = v2 = 0
    beta1, beta2 = 0.9, 0.999
    eps = 1e-8

    for t in range(1, epochs+1):
        z1, a1, z2, a2 = forward(X, W1, b1, W2, b2)
        dW1, db1, dW2, db2 = backward(X, y, z1, a1, z2, a2, W2)

        m1 = beta1*m1 + (1-beta1)*dW1
        v1 = beta2*v1 + (1-beta2)*(dW1**2)

        m1_hat = m1/(1-beta1**t)
        v1_hat = v1/(1-beta2**t)

        W1 -= lr*m1_hat/(np.sqrt(v1_hat)+eps)

    return W1, b1, W2, b2
```
